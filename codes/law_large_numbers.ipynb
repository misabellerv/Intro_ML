Hi guys! This notebook aims in experimental verification of a famous theorem in statistic: Law of large numbers.

In short words, this theorem says that the more times we perform an experiment, the smaller the uncertainty associated with it, that is, the smaller the variance. The arithmetic mean of the results of performing the same experiment over and over again tends to approach the expected value as more attempts are made. In other words, the more trials that are performed, the more the arithmetic mean probability of the observed results will approach the actual probability. Let's check if this is true. (Well, of course it's true, but since I'm a scientist, it's important to verify stuff...)

## Functions

So, I define a number of samples and a number of trials performed in experiment having more data avaliable. For each sample, I calculate the mean and then store them into a list and plot it's entire histogram. I'm checking this for N random variables that follow a Gaussian distribuiton.

```python
def gerador(n_samples, N):
    samples = []
    for i in range(n_samples):
        samples.append(np.random.normal(size=N))
    medias = []
    for i in samples:
        medias.append(np.mean(i))
    plt.hist(medias,bins=int(180/5),edgecolor = 'black')
    plt.title(f'Histogram - gaussian variables,N= {N}, n_samples={n_samples}')
    plt.grid(True)
    plt.xlabel('means')
```

## Experiment

Let's try for $N=1$

```python
gerador(10000,1)
```

We can see that the Normal distribuition is spread all over in the range of $(-4,+4)$

Let's try for $N = 10$

```python
gerador(10000,10)
```

Now, the gaussian range is reduced! The new x-axis range is between -1.0 and +1.0

Let's try for N=100

```python
gerador(10000,100)
```

Woah! The new range is $(-0.4,+0.4)$, that is, 10 times smaller than the initial range. We have the same samples number, but as we go increasing N, variance of normal distribuiton reduces and we aproximate
