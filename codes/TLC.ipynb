Hello! So, what if now we verify Central limit theorem? Sounds cool :-)

What is CLT? The central limit theorem is a famous statistics theorem that show distributions of sample means approximates a normal distribution as the sample size gets bigger regardless of the population's distribution. Can we check this? Yes! 
In this experiment I will show how this happens for 2 specific distribuitions: uniform and bernoulli. It's expected, from the theorem, that as their sample size gets bigger, sample's distribuiton approximates a normal one. That is a powerfull theorem. 

# Importing bibs

```python
import numpy.random as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
from scipy.stats import bernoulli
```

# Uniform distribuiton

We define a populattion described by uniform distribuition, and random sample means. As we increase sample size, an element of sample means is updated in terms of population in terms of sample sizes. Just look at loop for i in range(0,1000). We are looping this operation among all the population elements. For each element of sample size, we can plot a histogram of the means counts calculated inside the loop. Let's see how this looks like.

```python
population = np.uniform(0,1, size = 1000)

sample_means = np.rand(1000)

var = np.rand(1000)
fig, axs = plt.subplots(1, 4)
fig.set_size_inches(25, 5)
ind = 0

for ax in axs.flat:
    sizes = [1,5,10,1000]
    for i in range(0,1000):
        var = np.randint(1,1000,sizes[ind])
        sample_means[i] = population[var].mean()
    ax.hist(sample_means,bins=int(180/5),edgecolor = 'black')
    ax.grid(True)
    ax.set_title(f'Histogram uniform variables - Sample size = {sizes[ind]} ')
    ax.set_xlabel('means')
    
    
    ind += 1
```

As we can see, for sample size = 1, distribuitions of means are not very looking like something Normal...although for sample size = 1000, it gets more and more like a Gaussian!

## Bernoulli Distribuition

Now, the same proceedure is adopted, with only the difference that population set is generated by a bernoulli distribuiton. Let's see how this looks like.

```python
population = bernoulli.rvs(0.5,size=1000)

sample_means = np.rand(1000)

var = np.rand(1000)
fig, axs = plt.subplots(1, 4)
fig.set_size_inches(25, 5)
ind = 0

for ax in axs.flat:
    sizes = [1,5,10,1000]
    for i in range(0,1000):
        var = np.randint(1,1000,sizes[ind])
        sample_means[i] = population[var].mean()
    ax.hist(sample_means,bins=int(180/5),edgecolor = 'black')
    ax.grid(True)
    ax.set_title(f'Histogram bernoulli variables - Sample size = {sizes[ind]} ')
    ax.set_xlabel('means')
    
    ind += 1
```
